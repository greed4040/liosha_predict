```python
# https://machinelearningmastery.com/how-to-make-classification-and-regression-predictions-for-deep-learning-models-in-keras/
# example making new class predictions for a classification problem
from keras.models import Sequential
from keras.layers import Dense
from sklearn.datasets.samples_generator import make_blobs
from sklearn.preprocessing import MinMaxScaler

def work():
    # generate 2d classification dataset
    X, y = make_blobs(n_samples=100, centers=2, n_features=2, random_state=1)
    scalar = MinMaxScaler()
    scalar.fit(X)
    X = scalar.transform(X)
    # define and fit the final model
    model = Sequential()
    model.add(Dense(4, input_dim=2, activation='relu'))
    model.add(Dense(4, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam')
    model.fit(X, y, epochs=500, verbose=0)
    # new instances where we do not know the answer
    Xnew, _ = make_blobs(n_samples=3, centers=2, n_features=2, random_state=1)
    Xnew = scalar.transform(Xnew)
    # make a prediction
    ynew = model.predict_classes(Xnew)
    # show the inputs and predicted outputs
    for i in range(len(Xnew)):
        print("X=%s, Predicted=%s" % (Xnew[i], ynew[i]))
```

```python
from random import randint
from numpy import array
from numpy import argmax
from pandas import concat
from pandas import DataFrame
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense

# one hot encode sequence
def one_hot_encode(sequence, n_unique=10):
	encoding = list()
	for value in sequence:
		vector = [0 for _ in range(n_unique)]
		vector[value] = 1
		encoding.append(vector)
	return array(encoding)

# decode a one hot encoded string
def one_hot_decode(encoded_seq):
	return [argmax(vector) for vector in encoded_seq]

def generate(i):
    ln = 20
    a0=X_base[i+0:i+0+ln]
    a1=X_base[i+1:i+1+ln]
    a2=X_base[i+2:i+2+ln]
    a3=X_base[i+3:i+3+ln]
    a4=X_base[i+4:i+4+ln]
    
    a0 = a0.astype(int)
    a1 = a1.astype(int)
    a2 = a2.astype(int)
    a3 = a3.astype(int)
    a4 = a4.astype(int)
    
    e0=DataFrame(one_hot_encode(a0.tolist()))
    e1=DataFrame(one_hot_encode(a1.tolist()))
    e2=DataFrame(one_hot_encode(a2.tolist()))
    e3=DataFrame(one_hot_encode(a3.tolist()))
    e4=DataFrame(one_hot_encode(a4.tolist()))
    ea=concat([e0, e1, e2, e3, e4], axis=1)
    ea=ea.values
    # convert to 3d for input
    X = ea.reshape( len(ea), 5, 10 )
    ##print(a0, a1, a2, a3, a4)
    y = one_hot_encode(a4)
    return X, y
    #return None, None
```

```python
# define model
model = Sequential()
model.add(LSTM(100, batch_input_shape=(5, 5, 10), stateful=True))
model.add(Dense(10, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
```

```python
# fit model
for z in range(40):
    for i in range(180):
        print(z, i)
        X, y = generate(i)
        #print('decode',one_hot_decode(X[0]))
        #print('decode',one_hot_decode(X[1]))
        #print('decode',one_hot_decode(X[2]))
        #print('decode',one_hot_decode(X[3]))
        #print('decode',one_hot_decode(X[4]))
        #print('dec y:',one_hot_decode(y))

        model.fit(X, y, epochs=1, batch_size=5, verbose=2, shuffle=False)
        model.reset_states()
```

```python
# evaluate model on new data
X_e, y_e = generate(1)
yhat = model.predict(X_e, batch_size=5)
print('Expected:  %s' % one_hot_decode(y_e))
print('Predicted: %s' % one_hot_decode(yhat))
```
